```python
import cv2 as cv
import matplotlib.pyplot as plt
import sys
```

```python
img1 = cv.imread("../images/image1.jpeg")
img2 = cv.imread("../images/image2.jpeg")
img3 = cv.imread("../images/image3.jpeg")
img4 = cv.imread("../images/image4.jpeg")

if img1 is None:
    sys.exit("Error in loading image")

if img2 is None:
    sys.exit("Error in loading image")

if img3 is None:
    sys.exit("Error in loading image")

if img4 is None:
    sys.exit("Error in loading image")
```

```python
# images reading using cv is displayed in BGR format
# so we use a function to convert it into RGB format
```

```python
img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2,cv.COLOR_BGR2RGB)
img3 = cv.cvtColor(img3,cv.COLOR_BGR2RGB)
img4 = cv.cvtColor(img4,cv.COLOR_BGR2RGB)
```

```python
plt.subplot(2,2,1)
plt.imshow(img1)

plt.subplot(2,2,2)
plt.imshow(img2)

plt.subplot(2,2,3)
plt.imshow(img3)

plt.subplot(2,2,4)
plt.imshow(img4)
```

```python
# to make the images greyscale we take the pixel values and average them out
# an image has three channels (RGB) so we choose all the pixels in those three channels
#img.shape prints out (height, width, channels)
```

```python
(w,h) = img1.shape[0:2] #not including 2 -> channel number
for i in range(w):
    for j in range(h):
        (r,g,b) = img1[i,j]
        img1[i,j] = 0.21*r + 0.72*g + 0.07*b

(w,h) = img2.shape[0:2]
for i in range(w):
    for j in range(h):
        (r,g,b) = img2[i,j]
        img2[i,j] = 0.21*r + 0.72*g + 0.07*b

(w,h) = img3.shape[0:2]
for i in range(w):
    for j in range(h):
        (r,g,b) = img3[i,j]
        img3[i,j] = 0.21*r + 0.72*g + 0.07*b

(w,h) = img4.shape[0:2]
for i in range(w):
    for j in range(h):
        (r,g,b) = img4[i,j]
        img4[i,j] = 0.21*r + 0.72*g + 0.07*b
```

```python
plt.subplot(2,2,1)
plt.imshow(img1)

plt.subplot(2,2,2)
plt.imshow(img2)

plt.subplot(2,2,3)
plt.imshow(img3)

plt.subplot(2,2,4)
plt.imshow(img4)
```

```python
import os

os.chdir("../images") #change directory to 'images'
cv.imwrite("gimg1.jpeg",img1)
cv.imwrite("gimg2.jpeg",img2)
cv.imwrite("gimg3.jpeg",img3)
cv.imwrite("gimg4.jpeg",img4)
```

```python
cam = cv.VideoCapture(0)
cv.namedWindow("capture")
while True:
    result, image = cam.read()
    if not result:
         print("Error in capturing image")
         break

    cv.imshow("capture",image)
    k = cv.waitKey(1)
    if k%256==27: #ESC key means break the window
        break
    elif k%256==32: #space key means capture and save image
        cv.imwrite("webcam1.png",image)

cv.destroyAllWindows()
cam.release()
```
