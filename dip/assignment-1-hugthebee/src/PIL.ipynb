```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

img1 = Image.open("../images/image1.jpeg")
img2 = Image.open("../images/image2.jpeg")
img3 = Image.open("../images/image3.jpeg")
img4 = Image.open("../images/image4.jpeg")

plt.subplot(2,2,1)
plt.imshow(img1)

plt.subplot(2,2,2)
plt.imshow(img2)

plt.subplot(2,2,3)
plt.imshow(img3)

plt.subplot(2,2,4)
plt.imshow(img4)
```

```python
pixel = img1.load() #loads the pixel map
(w,h) = img1.size
for i in range(w):
    for j in range(h):
        (r,g,b) = pixel[i,j]
        arr = np.array([0.21*r + 0.72*g + 0.07*b, 0.21*r + 0.72*g + 0.07*b, 0.21*r + 0.72*g + 0.07*b])
        arr = arr.astype(int)
        pixel[i,j] = (arr[0],arr[1],arr[2])

pixel = img2.load() #loads the pixel map
(w,h) = img2.size
for i in range(w):
    for j in range(h):
        (r,g,b) = pixel[i,j]
        arr = np.array([0.21*r + 0.72*g + 0.07*b, 0.21*r + 0.72*g + 0.07*b, 0.21*r + 0.72*g + 0.07*b])
        arr = arr.astype(int)
        pixel[i,j] = (arr[0],arr[1],arr[2])

pixel = img3.load() #loads the pixel map
(w,h) = img3.size
for i in range(w):
    for j in range(h):
        (r,g,b) = pixel[i,j]
        arr = np.array([0.21*r + 0.72*g + 0.07*b, 0.21*r + 0.72*g + 0.07*b, 0.21*r + 0.72*g + 0.07*b])
        arr = arr.astype(int)
        pixel[i,j] = (arr[0],arr[1],arr[2])

pixel = img4.load() #loads the pixel map
(w,h) = img4.size
for i in range(w):
    for j in range(h):
        (r,g,b) = pixel[i,j]
        arr = np.array([0.21*r + 0.72*g + 0.07*b, 0.21*r + 0.72*g + 0.07*b, 0.21*r + 0.72*g + 0.07*b])
        arr = arr.astype(int)
        pixel[i,j] = (arr[0],arr[1],arr[2])
```
